```python
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cdflib
%matplotlib inline
```

```python
def test_context(data):
    with cdflib.CDF(fname) as cdf:
        cdf.cdf_info()
        return cdflib.CDF(fname)[data]

def interp_single(data_i):
    DATA_large = test_context(data_i)
    DATA_interp = np.interp(np.arange(0, len(DATA_large), 1265.15), np.arange(0, len(DATA_large)), DATA_large)
    return DATA_interp

def interp_multiple(data_i, data_j):
    DATA_large = test_context(data_i)[:,data_j]
    DATA_interp = np.interp(np.arange(0, len(DATA_large), 1265.15), np.arange(0, len(DATA_large)), DATA_large)
    return DATA_interp

def interp_single_2(data_i):
    
    DATA_large = test_context(data_i)
    DATA_filter = np.zeros([DATA_large.shape[0]])
            
    for i in range(0, DATA_large.shape[0]):
        if (abs(DATA_large[i])>25):
            DATA_filter[i] = 0
        else:
            DATA_filter[i] = DATA_large[i]
    
    DATA_interp = np.interp(np.arange(0, len(DATA_filter), 117.8), np.arange(0, len(DATA_filter)), DATA_filter)
    return DATA_interp

def interp_multiple_2(data_i, data_j):
    DATA_large = test_context(data_i)[:,data_j]
    DATA_filter = np.zeros([DATA_large.shape[0]])
            
    for i in range(0, DATA_large.shape[0]):
        if (abs(DATA_large[i])>10):
            DATA_filter[i] = 0
        else:
            DATA_filter[i] = DATA_large[i]
    
    DATA_interp = np.interp(np.arange(0, len(DATA_filter), 117.8), np.arange(0, len(DATA_filter)), DATA_filter)
    return DATA_interp
```

```python
EPOCH, BF1 = [], []
BGSM_X, BGSM_Y, BGSM_Z = [], [], []
BGSE_X, BGSE_Y, BGSE_Z = [], [], []

for i in range(1,10):
    fname = ("./wi_h2_mfi_2022070"+str(i)+"_v04.cdf")
    EPOCH = np.concatenate((EPOCH, interp_single(0)), axis=0)
    BF1 = np.concatenate((BF1, interp_single(2)), axis=0)
    BGSM_X = np.concatenate((BGSM_X, interp_multiple(3,0)), axis=0)
    BGSM_Y = np.concatenate((BGSM_Y, interp_multiple(3,1)), axis=0)
    BGSM_Z = np.concatenate((BGSM_Z, interp_multiple(3,2)), axis=0)
    BGSE_X = np.concatenate((BGSE_X, interp_multiple(4,0)), axis=0)
    BGSE_Y = np.concatenate((BGSE_Y, interp_multiple(4,1)), axis=0)
    BGSE_Z = np.concatenate((BGSE_Z, interp_multiple(4,2)), axis=0)

for i in range(11,32):
    fname = ("./wi_h2_mfi_202207"+str(i)+"_v04.cdf")
    EPOCH = np.concatenate((EPOCH, interp_single(0)), axis=0)
    
    BF1 = np.concatenate((BF1, interp_single(2)), axis=0)
    BGSM_X = np.concatenate((BGSM_X, interp_multiple(3,0)), axis=0)
    BGSM_Y = np.concatenate((BGSM_Y, interp_multiple(3,1)), axis=0)
    BGSM_Z = np.concatenate((BGSM_Z, interp_multiple(3,2)), axis=0)
    BGSE_X = np.concatenate((BGSE_X, interp_multiple(4,0)), axis=0)
    BGSE_Y = np.concatenate((BGSE_Y, interp_multiple(4,1)), axis=0)
    BGSE_Z = np.concatenate((BGSE_Z, interp_multiple(4,2)), axis=0)
    
data_wind = np.c_[np.transpose(EPOCH),
                  np.transpose(BF1),
                  np.transpose(BGSM_X),
                  np.transpose(BGSM_Y),
                  np.transpose(BGSM_Z),
                  np.transpose(BGSE_X),
                  np.transpose(BGSE_Y),
                  np.transpose(BGSE_Z)]

data=pd.DataFrame(data_wind)
data.columns = ['EPOCH',
             'BF1',
             'BGSM(X)',
             'BGSM(Y)',
             'BGSM(Z)',
             'BGSE(X)',
             'BGSE(Y)',
             'BGSE(Z)']

data.to_csv("wind.csv")
data.head()
```
